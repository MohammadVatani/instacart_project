```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
import os

os.chdir('C:\\Users\\Admin\\DM_project') #I use this line to change my current work directory
```

```python
products_df = pd.read_csv('.//input//products.csv')
order_products__prior_df = pd.read_csv('./input//order_products__prior.csv')
```

```python
dataset = pd.merge(order_products__prior_df,products_df,how='inner',on='product_id')
```

we keep first 10000 orders to find assosiation rules

```python
dataset = dataset[dataset['order_id']<10000]
dataset['temp']=1
```

```python
def encoder(x):
    if x<=0 :
        return False
    else: 
        return True
df = dataset.groupby(['order_id','product_name'])['temp'].sum().unstack().fillna(0)
df = df.applymap(encoder)
```

```python
freq_itemsets = apriori(df,min_support=0.01,use_colnames=True)
freq_itemsets
```

```python
rules = association_rules(freq_itemsets,metric='lift',min_threshold=1)
rules.sort_values(by='confidence',ascending=False)
```
